# Read in the observational Ap data
This dataset comes from the ftp server provided by [GFZ Helmholtz Centre Potsdam](https://www.gfz-potsdam.de/en/kp-index/)

## Imports

```python
import pandas as pd
```

## Read in the data

```python
df = pd.read_csv('../data/Kp_ap_Ap_SN_F107_2021.txt', skiprows=39, header=0, delim_whitespace=True)
```

```python
df
```

## Fix the multiple date time column issue
Use the `df.apply()` method to combine columns, and assign to a new datetime column, changing the index

```python
cols=["YYY","MM","DD"]
df['date'] = df[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
df.index = pd.to_datetime(df.date)
```

## Plot the Ap values

```python
df.Ap.plot()
```
